Построим набор моделей для значений random_state от 0 до 9. Получите model.score для каждого варианта.

```python
for i in range(10):
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state = i)
    model = LogisticRegression()
    model.fit(X_train, Y_train)
    print("{}: {}".format(i, model.score(X_test, Y_test)))
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score( model, X_train, Y_train, cv = 7)
scores.mean(), scores
```

```python
import numpy as np
from sklearn.model_selection import KFold

kf = KFold(n_splits=9)

scores = cross_val_score( model, X_train, Y_train, cv = kf)
scores.mean(), scores
```

```python
from sklearn.model_selection import StratifiedKFold
skf = StratifiedKFold(n_splits=9)

scores = cross_val_score( model, X_train, Y_train, cv = skf)
scores.mean(), scores
```

```python
tp = 0 # True positive
fp = 0 # False positive
fn = 0 # False negative
tn = 0 # True negative
import matplotlib.pyplot as plt

fpr1 = []
tpr1 = []
pr1 = []
re1 = []
predictions = model.predict_proba( X_test )
for i in np.arange(0.1,1,0.0001):
    for predicted_prob, actual in zip( predictions[:, 1], Y_test ):
        if predicted_prob >= i:
            predicted = 1
        else:
            predicted = 0
    
        if predicted == 1:
            if actual == 1:
                tp += 1
            else:
                fp += 1
    
        else:
            if actual == 1:
                fn += 1
            else:
                tn += 1
    fpr2 = fp / (fp + tn) #показатель ложных положительных результатов 
    tpr2 = tp / (tp + fn) #показатель истино положительных результатов
    precision  = tp / (tp + fp) #Precision − доля правильно предсказанных средипричисленных моделью к категории1
    recall = tp / (tp + fn) #Recall − доля правильно предсказанных среди категории1
    fpr1.append(fpr2)
    tpr1.append(tpr2)
    pr1.append(precision)
    re1.append(recall)

plt.plot( fpr1, tpr1 )  # посчитаем ROC-кривую (Receiver Operating Characteristic)
```

```python
#Используем готовые библиотеки
from sklearn.metrics import roc_curve
fpr1, tpr1, thres = roc_curve( Y_test, predictions[:, 1] )

plt.plot( fpr1, tpr1 )
```

## Построить график Precision - Recall на этих же данных.

```python
#строим график Precision - Recall на этих же данных
tp1 = 0 # True positive
fp1 = 0 # False positive
fn1 = 0 # False negative
tn1 = 0  # True negative

pr1 = []
re1 = []
thres1 = []
predictions = model.predict_proba( X_test )
for i in np.arange(0,1,0.1):
    for predicted_prob, actual in zip( predictions[:, 1], Y_test ):
        if predicted_prob >= i:
            predicted = 1
        else:
            predicted = 0
    
        if predicted == 1:
            if actual == 1:
                tp1 += 1
            else:
                fp1 += 1
    
        else:
            if actual == 1:
                fn1 += 1
            else:
                tn1 += 1
    precision  = tp1 / (tp1 + fp1) #Precision − доля правильно предсказанных средипричисленных моделью к категории1
    recall = tp1 / (tp1 + fn1) #Recall − доля правильно предсказанных среди категории1
    pr1.append(precision)
    re1.append(recall)
    thres1.append(i)
for line in zip( pr1, re1, thres1):
    print( line )
```

```python
plt.plot(re1, pr1)# строим график Precision - Recall на этих же данных
```

```python
#Используем готовые библиотеки
from sklearn.metrics import precision_recall_curve
pr, re, thres = precision_recall_curve( Y_test, predictions[:, 1] )
plt.plot(re, pr)
```
